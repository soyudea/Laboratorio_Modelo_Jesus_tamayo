```python
rango = np.linspace(-2*math.pi, 2*math.pi,100)
def g_x(x):
  return math.sin(x)

def f_x(x):
  return g_x(x**2)

lista = list(map(f_x,rango))  
plt.plot(rango,lista)
rango
plt.xlabel("Angulos [-360,369]")
plt.ylabel("g(f(x)) fx = x^2 g(x) = sin(x)")
plt.title("Función compuesta")
```
